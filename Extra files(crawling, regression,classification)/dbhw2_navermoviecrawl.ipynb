```python
import requests
import sqlite3
conn=sqlite3.connect('hw.db')
cur=conn.cursor()
cur.execute("CREATE TABLE test1(number text, movie text, contents text, id text)")

import numpy as np
import pandas as pd
import requests
import urllib
from bs4 import BeautifulSoup

```

```python
temp=[]
```

```python
def playin(pagenum):
    url='https://movie.naver.com/movie/point/af/list.nhn'
    params={'page':int(pagenum)}
    res=requests.get(url,params=params)
    navigator=BeautifulSoup(res.text,'html.parser',from_encoding='utf-8')
    table=navigator.find('table',class_='list_netizen')
    
    global temp
    for i, r in enumerate(table.find_all('tr')):
        for j,c in enumerate(r.find_all('td')):
            if j==0:
                record1=(c.text.strip())
            elif j==1:
                record2=str(c.find('a').text.strip())
                record3=str(c.text.strip().split('\t')[0].split('\n')[1])
            elif j==2:
                record4=str(c.find('a').text.strip())
        try:
            if(record3!=''):
                temp.append((record1,record2,record3,record4))
        except:
            pass
            
    
```

```python
for i in range(1,130,1):
    playin(i)
```

```python
sql_query="INSERT INTO test1 VALUES(?,?,?,?)"
cur.executemany(sql_query,temp)
```
